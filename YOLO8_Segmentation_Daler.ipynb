```python

import os

from ultralytics import YOLO

data_path = '/content/config.yaml'

n_epochs = 1000
bs = 25
gpu_id = 0
verbose = True
rng = 0
validate = True
# 6
model = YOLO('/content/gdrive/MyDrive/Smirnova/runs/segment/train9/weights/last.pt')


results = model.train(
    data=data_path,
    epochs=n_epochs,
    batch=bs,
    device=gpu_id,
    verbose=verbose,
    seed=rng,
    val=validate,
    patience=0,

)
```
